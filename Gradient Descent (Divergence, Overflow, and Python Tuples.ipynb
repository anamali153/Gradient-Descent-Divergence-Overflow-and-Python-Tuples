# Using A Function 

# $$g(x) = x^5 - 2x^4 + 2$$

```python
#Make some data

x_3 = np.linspace(start = -2, stop=2, num=1000)



def h(x):
    return x**5 - 2*x**4 + 2



def dh(x):
    return 5*x**4 - 8*x**3
```

```python
#Chart One Cost Function
plt.figure(figsize= [15,5])


plt.subplot(1,2,1)


plt.xlim(-1, 4)
plt.ylim(0.0004, 3)

plt.title('Cost Function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('h(x)', fontsize=16)

plt.plot(x_3, h(x_3), color='blue', linewidth=3)


#second chart Derivative

plt.subplot(1,2,2)


plt.title('Slope Of The Cost Function', fontsize=17)
plt.xlabel('X', fontsize=16)
plt.ylabel('dh(x)', fontsize=16)
plt.grid()
plt.xlim(-2, 3)
plt.ylim(-3, 6)

plt.plot(x_3, dh(x_3), color='skyblue', linewidth=3)





plt.show()
```

# Gradient Descent As Python Function

```python
def gradient_descent(derivative_func, initial_guess,multiplier, precision):


    new_x = initial_guess        #(starting point)
    x_list = [new_x]
    slope_list= [derivative_func(new_x)]

    for n in range(500):
        previous_x= new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)
       # print(step_size)  

        x_list.append(new_x)
        slope_list.append(derivative_func(new_x))


        if step_size < precision:
            break

    return new_x, x_list, slope_list
```
